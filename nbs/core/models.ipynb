# models

> Data models for scrollbar configuration, state, and HTML ID generation.

```python
#| default_exp core.models
```

```python
#| export
from dataclasses import dataclass
from typing import Optional
```

## ScrollbarConfig

```python
#| export
@dataclass
class ScrollbarConfig:
    """Initialization-time configuration for a virtual scrollbar."""
    prefix: str = "sb"                  # HTML ID prefix
    show_scrollbar: bool = True          # Enable/disable scrollbar rendering
    min_thumb_height: int = 24           # Minimum thumb height in pixels
    track_width: int = 3                 # Track width in Tailwind units (w-3 = 12px)
    auto_hide: bool = True               # Hide when total_items <= visible_count
```

## ScrollbarState

```python
#| export
@dataclass
class ScrollbarState:
    """Mutable runtime state for a virtual scrollbar."""
    position: int = 0                              # Current position (0-indexed)
    visible_count: int = 1                         # Number of visible items
    total_items: int = 0                           # Total item count
    max_position: Optional[int] = None             # Upper bound of position range (None = total_items - visible_count)
    thumb_ratio: Optional[float] = None            # Thumb height as fraction of track (None = visible_count / total_items)
```

## ScrollbarIds

```python
#| export
@dataclass
class ScrollbarIds:
    """HTML element ID generators for a scrollbar instance."""
    prefix: str = "sb"  # Instance prefix

    @property
    def track(self) -> str: return f"{self.prefix}-scrollbar-track"  # Scrollbar track container

    @property
    def thumb(self) -> str: return f"{self.prefix}-scrollbar-thumb"  # Draggable thumb element
```

## Tests

```python
# Default config
cfg = ScrollbarConfig()
assert cfg.prefix == "sb"
assert cfg.show_scrollbar is True
assert cfg.min_thumb_height == 24
assert cfg.track_width == 3
```

```python
# Default state
state = ScrollbarState()
assert state.position == 0
assert state.visible_count == 1
assert state.total_items == 0
```

```python
# ID generation
ids = ScrollbarIds(prefix="cs0")
assert ids.track == "cs0-scrollbar-track"
assert ids.thumb == "cs0-scrollbar-thumb"
```
